```python
import networkx as nx
import matplotlib as plt
import numpy as np
```

```python
def CircularGraph(n) :
    G=nx.Graph()
    for i in range(n):
        G.add_node(i)
    for i in range(n):
        G.add_edge(i, (i+1)%n)
    return G

def CompleteBipartiteGraph(n, m) :
    G=nx.Graph()
    for i in range(n+m) :
        G.add_node(i)
    for i in range(n) :
        for j in range(m) :
            G.add_edge(i, j+n)
    return G

# Ui(외부) - Vi(내부) Adjacent
# Vi(내부) - V(i+k)modN Adjacent
# General Peterson Graph
def GeneralizedPetersonGraph(n, k) :
    G=nx.Graph()
    for i in range(2*n) :
        G.add_node(i)
    for i in range(n) :
        G.add_edge(i , (i+k)%n)
        G.add_edge(i, i+n)
        G.add_edge(n+i, (1+i)%n +n)
    return G

# Digraph가 주어졌을 때 a - b 까지의 minimum cost구하기
def DA(G, a, b) :
    path = nx.dijkstra_path(G, a, b)
    value = 0
    for i in range(len(path)-1):
        value += G.get_edge_data(path[i], path[i+1])['weight']
    return value
```

```python
circularG = CircularGraph(7)
nx.draw(circularG, with_labels=1)
```

```python
completeBipartiteG = CompleteBipartiteGraph(2, 2)
nx.draw(completeBipartiteG, with_labels=1)
```

```python
generalizedPetersonG = GeneralizedPetersonGraph(5, 1)
nx.draw(generalizedPetersonG, with_labels=1)
```

```python
X=nx.DiGraph()
for i in range(5) :
    X.add_node(i+1)
X.add_edge(1, 2, weight=70)
X.add_edge(1, 4, weight=50)
X.add_edge(1, 3, weight=30)
X.add_edge(2, 3, weight=40)
X.add_edge(2, 4, weight=90)
X.add_edge(1, 5, weight=100)

nx.draw(X, with_labels=1)

print(DA(X, 1, 4))
```
