```python
# import: networkx, numpy, matplotlib
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Cycel Graph 생성

S = nx.cycle_graph(4)
nx.draw(S, with_labels=1)
```

```python
# Complete Multipartite Graph 생성
## 0,1 / 2,3,4 / 5,6,7 이 각 partition
## 0 ,,, n-1 로 vertex index가 고정이 됨

A = nx.complete_multipartite_graph(2, 3, 3)
nx.draw(A, with_labels=1)
```

```python
# Bipartite Graph의 생성
B = nx.Graph()

node_set_1 = ['A', 'B', 'C']
node_set_2 = ['a', 'b', 'c']

for node in node_set_1 :
    B.add_node(node)
    
for node in node_set_2 :
    B.add_node(node)
    
for node1 in node_set_1 :
    for node2 in node_set_2 : 
        B.add_edge(node1, node2)
        
nx.draw(B, with_labels=1)
```

```python
# Complete Bipartite Graph 생성
C=nx.complete_bipartite_graph(3, 3)

nx.draw(C, with_labels=1)
```

```python
## Vertex의 추가, Edge의 추가
C.add_node(6)
C.add_edge(6, 3)

nx.draw(C, with_labels=1)
```

```python
# Digraph의 생성
D = nx.DiGraph()

node_set = ['A', 'B', 'C', 'D', 'E']

## Vertex 추가
for node in node_set :
    D.add_node(node)
    
## Edge 추가
D.add_edge('A', 'B', weight=10)
D.add_edge('B', 'C', weight=20)
D.add_edge('C', 'D', weight=30)
D.add_edge('A', 'E', weight=40)
D.add_edge('B', 'D', weight=1000)

## Graph 도시
nx.draw(D, with_labels=1)
```

```python
# NetworkX 사용한 Dijkstra 적용
path = nx.dijkstra_path(D, 'A', 'D')
print(path)
```

```python
# edge weight 알아내기
D.get_edge_data('B', 'D')
```

```python
# Dijkstra 적용한 뒤에 shortest path distance 구하기
value = 0
for v in range(len(path)-1):
    value += D.get_edge_data(path[v], path[v+1])['weight']
print(value)
```
